# Importing The Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

# Data Preprocessing

```python
df = pd.read_csv("../Data/Raw/fitness_tracker_dataset.csv")
df
```

**Columns In The Dataset**

```python
print("Total Columns in The Dataset")
print(df.columns)
print("\n")
```

**NUll Values In The Columns**

```python
print("Column-wise Null value")
print(df.isnull().sum())
print('\n')
```

**DataTypes Of The Columns**

```python
print("Data Types of All the Columns ")
print(df.dtypes)
```

**Data Cleaning**

***Deleting The Null Values***

```python
dataFrame = df.dropna().reset_index(drop=True)
dataFrame
```

***Deleting Columns That Are Not Needed***

```python
dataFrame = dataFrame.drop(columns = ["user_id", "location", "active_minutes"])
dataFrame
```

***Converting 'date' Column into DateTime Data type***

```python
dataFrame['date'] = pd.to_datetime(dataFrame['date'])
print(dataFrame['date'].dtype)
```

**Splitting The Dataset Based On The Year**

```python
dataFrame['year'] = dataFrame['date'].dt.year
df_by_year = {year: dataFrame[dataFrame['year'] == year].drop(columns='year') for year in dataFrame['year'].unique()}

df_2023 = df_by_year[2023]
df_2024 = df_by_year[2024]
df_2025 = df_by_year[2025]
```

```python
print("Year : 2023")
df_2023
```

```python
print("Year : 2024")
df_2024
```

```python
print("Year : 2025")
df_2025
```

**Store Into Different CSV Files**

```python
df_arr = {
    '2023': df_2023,
    '2024': df_2024,
    '2025': df_2025  
}

directory = '../Data/Processed'
if not os.path.exists(directory):
    os.makedirs(directory)

for year, df in df_arr.items():
    year_directory = os.path.join(directory, year)
    
    if not os.path.exists(year_directory):
        os.makedirs(year_directory)
    
    filename = os.path.join(year_directory, f'data_{year}.csv')
    
    df.to_csv(filename, index=False)
    print(f'Saved {filename}')
```
